```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

T_vdw = 0.9
fig, ax = plt.subplots()
marker = ["*", "o", "^"]

for i in range(3):
    V_vdw = np.linspace(0.45, 2.45, 20)
    P_vdw = 8*T_vdw/(3*V_vdw-1)-3/(V_vdw**2)
    T_ig = T_vdw
    V_ig = np.linspace(0.45, 2.45, 20)
    P_ig = 8*T_vdw/(3*V_ig)
    
    ax.plot(V_vdw, P_vdw, '-{}b'.format(marker[i]), label='$T$={}$T_c$'.format(T_vdw))
    ax.plot(V_ig, P_ig, '-{}g'.format(marker[i]), label='$T$={}$T_c$'.format(T_vdw))
    T_vdw += 0.1

ax.set(xlabel='$V$ ($V_c$)', ylabel='$P$ ($P_c$)',
       title='PV curve for van der Waals gas and ideal gas')
ax.set_ylim([0,7])
ax.legend()
ax.grid()

fig.savefig("Problem1_3.png")
plt.show()
```
