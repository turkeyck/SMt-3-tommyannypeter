```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

T_vdw = 0.9
V_vdw = np.linspace(0.45, 4.45, 80)
P_vdw = 8*T_vdw/(3*V_vdw-1)-3/(V_vdw**2)

T_ig = T_vdw
V_ig = np.linspace(0.45, 4.45, 80)
P_ig = 8*T_vdw/(3*V_ig)

fig, ax = plt.subplots()
ax.plot(V_vdw, P_vdw, '-*b', label='$T$=0.9$T_c$')
ax.plot(V_ig, P_ig, '-*g', label='$T$=0.9$T_c$')

ax.set(xlabel='$V$ ($V_c$)', ylabel='$P$ ($P_c$)',
       title='PV curve for van der Waals gas and ideal gas')
ax.legend()
ax.grid()

fig.savefig("Problem1_2.png")
plt.show()
```
