```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

T_vdw = 0.9
V_vdw = np.linspace(0.58, 4.58, 80)
P_vdw = 8*T_vdw/(3*V_vdw-1)-3/(V_vdw**2)

T_ig = 1.0
V_ig = np.linspace(0.58, 4.58, 80)
P_ig = 1/V_ig

plt.plot(V_vdw, P_vdw, '-*b', label='T=0.9Tc')
plt.plot(V_ig, P_ig, '-*g', label='T')
plt.xlabel("$V$ ($V_c$)")
plt.ylabel("$P$ ($P_c$)")
plt.title("PV curve for van der Waals gas")

plt.savefig("Problem1_2.png")
plt.show()
```
